```python
#task2

def count_unread_messages(messages):
    total = messages["count"]

    for grup in messages["subgrup"]:
        total += count_unread_messages(grup)

    return total  
    
messages = {
    "count" : 10,
    "subgrup" : [
        {
            "count" : 5,
            "subgrup" : []
        }
        ,
        {
            "count" : 10,
            "subgrup" : [
                {
                    "count" : 20,
                    "subgrup" : []
                }
                ,
                {
                    "count" : 30,
                    "subgrup" : []
                }
            ]
        }
    ]
}

print(count_unread_messages(messages))            
```

```python
#task4

def format_number_short(n):
    if n < 1000:
        return str(n)
    elif n < 1000000:
        return str(round(n/1000,1)) + "k"
    else:
        return format_number_short(n/1000000) + "M"

n = int(input("enter a number")) 
format_number_short(n)   
```
