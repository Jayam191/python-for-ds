```python
#task4

follower =  [950, 1500, 25000, 1200000]

def format_follower(count):
    if count >= 1000000:
        return f"{count/1000000:.1f}M"
    elif count>=1000:
        return f"{count/1000:.1f}k"
    else:
        return str(count)
        
result = list(map(format_follower,follower))  
result
```
