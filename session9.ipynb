```python
#task1

def cal(price,discount):
    discount = price*discount/100
    return price-discount

cal(1000,20)
```

```python
#task3

def format_price(price, currency='INR'):
    if currency == "INR":
        print(f"₹{price}")
    else:
        print("$",price)
format_price(500,"usd")
```

```python
#task4

def apply_coupon(price, coupon_code=None):
    if coupon_code == "ZOMATO10":
        discount = price*10/100
        return price -discount
    else:
        return price

apply_coupon(100,"ZOMATO10")
```
